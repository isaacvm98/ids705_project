```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

from pygam import LinearGAM, s, l
from sklearn.metrics import mean_absolute_error, mean_squared_error

# =========================================================
# 1. Load data
# =========================================================
train = pd.read_csv("training.csv", parse_dates=["week_start"])
test = pd.read_csv("test.csv", parse_dates=["week_start"])

target = "weekly_units"
drop_cols = ["weekly_units", "week_start"]
feature_cols = [c for c in train.columns if c not in drop_cols]

X_train = train[feature_cols].copy()
y_train = train[target].copy()

X_test = test[feature_cols].copy()
y_test = test[target].copy()

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

# =========================================================
# 2. Define GAM structure
# =========================================================
smooth_cols = [
    "lag_1",
    "lag_2",
    "lag_4",
    "lag_52",
    "rolling_mean_4",
    "rolling_std_4",
    "rolling_mean_8",
    "momentum",
    "recent_cv",
    "time_index",
    "avg_price",
    "price_rolling_4",
    "price_relative",
    "store_traffic_lag1",
    "cat_share_lag1",
]

linear_cols = [c for c in feature_cols if c not in smooth_cols]
col_index = {col: i for i, col in enumerate(feature_cols)}

terms = None
for col in feature_cols:
    idx = col_index[col]
    term = s(idx) if col in smooth_cols else l(idx)
    terms = term if terms is None else terms + term

# =========================================================
# 3. Metric helpers
# =========================================================
def smape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    mask = denom != 0
    if mask.sum() == 0:
        return np.nan
    return 100 * np.mean(np.abs(y_true[mask] - y_pred[mask]) / denom[mask])

def mape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return 100 * np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask]))

def metric_table(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mean_sales = np.mean(y_true)
    return {
        "RMSE": rmse,
        "MAE": mae,
        "NRMSE_%_of_mean_sales": (100 * rmse / mean_sales) if mean_sales != 0 else np.nan,
        "MAE_%_of_mean_sales": (100 * mae / mean_sales) if mean_sales != 0 else np.nan,
        "MAPE": mape(y_true, y_pred),
        "sMAPE": smape(y_true, y_pred),
        "Mean_Sales": mean_sales,
        "n_weeks": len(y_true),
    }

# =========================================================
# 4. Fit tuned GAM
# =========================================================
best_lam = 10

gam_tuned = LinearGAM(terms, lam=best_lam)
gam_tuned.fit(X_train.values, y_train.values)

train_pred = gam_tuned.predict(X_train.values)
test_pred = gam_tuned.predict(X_test.values)

print(f"\nTuned GAM fitted with lam = {best_lam}")

print("\nTrain metrics")
train_metrics = metric_table(y_train, train_pred)
for k, v in train_metrics.items():
    if isinstance(v, (int, float, np.floating)):
        print(f"{k}: {v:.3f}")

print("\nTest metrics")
test_metrics = metric_table(y_test, test_pred)
for k, v in test_metrics.items():
    if isinstance(v, (int, float, np.floating)):
        print(f"{k}: {v:.3f}")

# =========================================================
# 5. Reconstruct store and category labels
# =========================================================
store_cols = [
    "store_id_CA_1", "store_id_CA_2", "store_id_CA_3", "store_id_CA_4",
    "store_id_TX_1", "store_id_TX_2", "store_id_TX_3",
    "store_id_WI_1", "store_id_WI_2", "store_id_WI_3",
]

def get_store(row):
    for c in store_cols:
        if row[c] == 1:
            return c.replace("store_id_", "")
    return "UNKNOWN"

def get_category(row):
    if row["cat_id_FOODS"] == 1:
        return "FOODS"
    elif row["cat_id_HOBBIES"] == 1:
        return "HOBBIES"
    elif row["cat_id_HOUSEHOLD"] == 1:
        return "HOUSEHOLD"
    return "UNKNOWN"

results = test.copy()
results["store_id"] = results.apply(get_store, axis=1)
results["category_id"] = results.apply(get_category, axis=1)
results["actual_units"] = y_test.values
results["predicted_units_raw"] = test_pred
results["predicted_units"] = np.rint(test_pred).astype(int)
results["predicted_units"] = results["predicted_units"].clip(lower=0)

# =========================================================
# 6. Export 1: prediction file
# =========================================================
predictions_clean = (
    results[["week_start", "store_id", "category_id", "actual_units", "predicted_units"]]
    .sort_values(["week_start", "store_id", "category_id"])
    .reset_index(drop=True)
)

predictions_clean.to_csv("gam_tuned_predictions_clean.csv", index=False)
print("\nSaved: gam_tuned_predictions_clean.csv")

# =========================================================
# 7. Export 2: overall metrics
# =========================================================
overall_metrics = pd.DataFrame([{
    "model": f"GAM (Tuned, lam={best_lam})",
    **metric_table(results["actual_units"], results["predicted_units_raw"])
}])

overall_metrics_export = overall_metrics[[
    "model",
    "MAE",
    "RMSE",
    "MAPE",
    "sMAPE",
    "NRMSE_%_of_mean_sales",
    "MAE_%_of_mean_sales",
    "Mean_Sales",
    "n_weeks",
]]

print("\nOVERALL METRICS")
display(overall_metrics_export.round(3))

overall_metrics_export.to_csv("gam_tuned_overall_metrics.csv", index=False)
print("Saved: gam_tuned_overall_metrics.csv")

# =========================================================
# 8. Export 3: category metrics
# =========================================================
category_metrics = (
    results.groupby("category_id", sort=True)
    .apply(lambda g: pd.Series(metric_table(g["actual_units"], g["predicted_units_raw"])))
    .reset_index()
    .rename(columns={"category_id": "Category"})
    .sort_values("Category")
    .reset_index(drop=True)
)

print("\nCATEGORY METRICS")
display(category_metrics.round(3))

category_metrics.to_csv("gam_tuned_category_metrics.csv", index=False)
print("Saved: gam_tuned_category_metrics.csv")

# =========================================================
# 9. Export 4: store-category metrics
# =========================================================
store_category_metrics = (
    results.groupby(["category_id", "store_id"], sort=True)
    .apply(lambda g: pd.Series(metric_table(g["actual_units"], g["predicted_units_raw"])))
    .reset_index()
    .rename(columns={"category_id": "Category", "store_id": "Store"})
    .sort_values(["Category", "Store"])
    .reset_index(drop=True)
)

print("\nSTORE-CATEGORY METRICS")
display(store_category_metrics.round(3))

store_category_metrics.to_csv("gam_tuned_store_category_metrics.csv", index=False)
print("Saved: gam_tuned_store_category_metrics.csv")

# =========================================================
# 10. Store-category summary
# =========================================================
store_cat_summary = pd.DataFrame([{
    "num_series": store_category_metrics.shape[0],
    "mean_MAE": store_category_metrics["MAE"].mean(),
    "median_MAE": store_category_metrics["MAE"].median(),
    "mean_RMSE": store_category_metrics["RMSE"].mean(),
    "median_RMSE": store_category_metrics["RMSE"].median(),
    "mean_MAPE": store_category_metrics["MAPE"].mean(),
    "median_MAPE": store_category_metrics["MAPE"].median(),
    "mean_sMAPE": store_category_metrics["sMAPE"].mean(),
    "median_sMAPE": store_category_metrics["sMAPE"].median(),
    "mean_NRMSE_%": store_category_metrics["NRMSE_%_of_mean_sales"].mean(),
    "median_NRMSE_%": store_category_metrics["NRMSE_%_of_mean_sales"].median(),
}])

print("\nSTORE-CATEGORY SUMMARY")
display(store_cat_summary.round(3))

store_cat_summary.to_csv("gam_tuned_store_category_summary.csv", index=False)
print("Saved: gam_tuned_store_category_summary.csv")

# =========================================================
# 11. Best and worst store-category series by sMAPE
# =========================================================
best_5 = store_category_metrics.nsmallest(5, "sMAPE")
worst_5 = store_category_metrics.nlargest(5, "sMAPE")

print("\nBEST 5 STORE-CATEGORY SERIES BY sMAPE")
display(best_5.round(3))

print("\nWORST 5 STORE-CATEGORY SERIES BY sMAPE")
display(worst_5.round(3))

best_5.to_csv("gam_tuned_best_5_store_category.csv", index=False)
worst_5.to_csv("gam_tuned_worst_5_store_category.csv", index=False)

# =========================================================
# 12. Consolidated Excel export
# =========================================================
with pd.ExcelWriter("gam_tuned_outputs.xlsx", engine="openpyxl") as writer:
    predictions_clean.to_excel(writer, sheet_name="Predictions", index=False)
    overall_metrics_export.to_excel(writer, sheet_name="Overall_Metrics", index=False)
    category_metrics.to_excel(writer, sheet_name="Category_Metrics", index=False)
    store_category_metrics.to_excel(writer, sheet_name="Store_Category_Metrics", index=False)
    store_cat_summary.to_excel(writer, sheet_name="Store_Category_Summary", index=False)
    best_5.to_excel(writer, sheet_name="Best_5_Series", index=False)
    worst_5.to_excel(writer, sheet_name="Worst_5_Series", index=False)

print("Saved: gam_tuned_outputs.xlsx")

# =========================================================
# 13. Top-5 interpretability plots for tuned GAM
# Rank smooth terms by partial dependence range
# =========================================================
plot_dir = Path("gam_tuned_top5_interpretability")
plot_dir.mkdir(exist_ok=True)

smooth_term_info = []

for i, term in enumerate(gam_tuned.terms):
    if term.isintercept:
        continue

    feature_idx = getattr(term, "feature", None)
    if feature_idx is None:
        continue

    feature_name = feature_cols[feature_idx]
    if feature_name not in smooth_cols:
        continue

    XX = gam_tuned.generate_X_grid(term=i)
    pdep = gam_tuned.partial_dependence(term=i, X=XX)

    effect_range = float(np.max(pdep) - np.min(pdep))

    smooth_term_info.append({
        "term_index": i,
        "feature": feature_name,
        "partial_dependence_range": effect_range
    })

top5_terms = (
    pd.DataFrame(smooth_term_info)
    .sort_values("partial_dependence_range", ascending=False)
    .head(5)
    .reset_index(drop=True)
)

print("\nTOP 5 SMOOTH TERMS")
display(top5_terms.round(3))

top5_terms.to_csv(plot_dir / "top5_smooth_terms.csv", index=False)

# Combined figure
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(10, 20))
axes = np.atleast_1d(axes)

for ax, row in zip(axes, top5_terms.to_dict("records")):
    i = row["term_index"]
    feature_name = row["feature"]
    feature_idx = feature_cols.index(feature_name)

    XX = gam_tuned.generate_X_grid(term=i)
    pdep = gam_tuned.partial_dependence(term=i, X=XX)

    ax.plot(XX[:, feature_idx], pdep)
    ax.set_title(
        f"{feature_name} | partial dependence range = {row['partial_dependence_range']:.2f}"
    )
    ax.set_xlabel(feature_name)
    ax.set_ylabel("Effect on weekly_units")

    # individual file
    plt.figure(figsize=(8, 4))
    plt.plot(XX[:, feature_idx], pdep)
    plt.title(
        f"{feature_name} | partial dependence range = {row['partial_dependence_range']:.2f}"
    )
    plt.xlabel(feature_name)
    plt.ylabel("Effect on weekly_units")
    plt.tight_layout()
    plt.savefig(plot_dir / f"{feature_name}_partial_dependence.png", dpi=200)
    plt.close()

plt.tight_layout()
plt.savefig(plot_dir / "top5_partial_dependence_combined.png", dpi=200)
plt.show()

print(f"\nSaved interpretability outputs in: {plot_dir}")

# =========================================================
# 14. Plain vanilla vs tuned comparison table
# =========================================================
plain_vanilla_metrics = {
    "model": "GAM (Plain Vanilla)",
    "MAE": 595.593,
    "RMSE": 957.190,
    "MAPE": 7.305,
    "sMAPE": 7.758,
}

tuned_metrics = {
    "model": f"GAM (Tuned, lam={best_lam})",
    "MAE": overall_metrics_export.loc[0, "MAE"],
    "RMSE": overall_metrics_export.loc[0, "RMSE"],
    "MAPE": overall_metrics_export.loc[0, "MAPE"],
    "sMAPE": overall_metrics_export.loc[0, "sMAPE"],
}

comparison_df = pd.DataFrame([plain_vanilla_metrics, tuned_metrics])

print("\nPLAIN VANILLA VS TUNED GAM")
display(comparison_df.round(3))

comparison_df.to_csv("gam_plain_vs_tuned_comparison.csv", index=False)
print("Saved: gam_plain_vs_tuned_comparison.csv")
```

```python
import pandas as pd
import numpy as np
from pygam import LinearGAM, s, l
from sklearn.metrics import mean_absolute_error, mean_squared_error

# =========================================================
# 1. Load data
# =========================================================
train = pd.read_csv("training.csv", parse_dates=["week_start"])
test = pd.read_csv("test.csv", parse_dates=["week_start"])

target = "weekly_units"

# =========================================================
# 2. Reconstruct store and category labels
# =========================================================
store_cols = [
    "store_id_CA_1", "store_id_CA_2", "store_id_CA_3", "store_id_CA_4",
    "store_id_TX_1", "store_id_TX_2", "store_id_TX_3",
    "store_id_WI_1", "store_id_WI_2", "store_id_WI_3",
]

def get_store(row):
    for c in store_cols:
        if row[c] == 1:
            return c.replace("store_id_", "")
    return "UNKNOWN"

def get_category(row):
    if row["cat_id_FOODS"] == 1:
        return "FOODS"
    elif row["cat_id_HOBBIES"] == 1:
        return "HOBBIES"
    elif row["cat_id_HOUSEHOLD"] == 1:
        return "HOUSEHOLD"
    return "UNKNOWN"

for df in [train, test]:
    df["store_id"] = df.apply(get_store, axis=1)
    df["category_id"] = df.apply(get_category, axis=1)
    df["store_category"] = df["store_id"] + "_" + df["category_id"]

# =========================================================
# 3. Find Pareto subset covering 80% of training sales
# =========================================================
sales_by_combo = (
    train.groupby("store_category")[target]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={target: "total_sales"})
)

sales_by_combo["cum_sales"] = sales_by_combo["total_sales"].cumsum()
sales_by_combo["cum_share"] = sales_by_combo["cum_sales"] / sales_by_combo["total_sales"].sum()

pareto_combos = sales_by_combo.loc[sales_by_combo["cum_share"] <= 0.80, "store_category"].tolist()

# include the first combo that crosses 80%, if needed
if len(pareto_combos) < len(sales_by_combo):
    next_combo = sales_by_combo.loc[len(pareto_combos), "store_category"]
    pareto_combos.append(next_combo)

print("Number of Pareto store-category combinations:", len(pareto_combos))
print("Pareto combinations:")
print(pareto_combos)

# =========================================================
# 4. Filter train and test to Pareto subset
# =========================================================
train_pareto = train[train["store_category"].isin(pareto_combos)].copy()
test_pareto = test[test["store_category"].isin(pareto_combos)].copy()

print("\nPareto train shape:", train_pareto.shape)
print("Pareto test shape :", test_pareto.shape)

# =========================================================
# 5. Define feature set
# =========================================================
drop_cols = [
    "weekly_units",
    "week_start",
    "store_id",
    "category_id",
    "store_category"
]
feature_cols = [c for c in train_pareto.columns if c not in drop_cols]

X_train = train_pareto[feature_cols].copy()
y_train = train_pareto[target].copy()

X_test = test_pareto[feature_cols].copy()
y_test = test_pareto[target].copy()

# =========================================================
# 6. Tuned GAM structure
# =========================================================
smooth_cols = [
    "lag_1",
    "lag_2",
    "lag_4",
    "lag_52",
    "rolling_mean_4",
    "rolling_std_4",
    "rolling_mean_8",
    "momentum",
    "recent_cv",
    "time_index",
    "avg_price",
    "price_rolling_4",
    "price_relative",
    "store_traffic_lag1",
    "cat_share_lag1",
]

col_index = {col: i for i, col in enumerate(feature_cols)}

terms = None
for col in feature_cols:
    idx = col_index[col]
    term = s(idx) if col in smooth_cols else l(idx)
    terms = term if terms is None else terms + term

best_lam = 10  # tuned raw-demand GAM

# =========================================================
# 7. Metrics
# =========================================================
def smape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    mask = denom != 0
    if mask.sum() == 0:
        return np.nan
    return 100 * np.mean(np.abs(y_true[mask] - y_pred[mask]) / denom[mask])

def mape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return 100 * np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask]))

def evaluate(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    return {
        "RMSE": rmse,
        "MAE": mae,
        "MAPE": mape(y_true, y_pred),
        "sMAPE": smape(y_true, y_pred),
    }

# =========================================================
# 8. Fit tuned GAM on Pareto subset
# =========================================================
gam_pareto = LinearGAM(terms, lam=best_lam)
gam_pareto.fit(X_train.values, y_train.values)

test_pred = gam_pareto.predict(X_test.values)
test_pred = np.clip(test_pred, a_min=0, a_max=None)

pareto_metrics = evaluate(y_test, test_pred)

print("\nTuned GAM on Pareto subset - Test metrics")
for k, v in pareto_metrics.items():
    print(f"{k}: {v:.3f}")

# =========================================================
# 9. Compare against full-data tuned GAM
# Update these if needed from your final results
# =========================================================
comparison_df = pd.DataFrame([
    {
        "model": "GAM (Tuned Raw Demand, Full Data)",
        "RMSE": 822.109,
        "MAPE": 4.597,
        "sMAPE": 4.598,
    },
    {
        "model": "GAM (Tuned Raw Demand, Pareto 80%)",
        "RMSE": pareto_metrics["RMSE"],
        "MAPE": pareto_metrics["MAPE"],
        "sMAPE": pareto_metrics["sMAPE"],
    }
])

print("\nFull vs Pareto comparison")
display(comparison_df.round(3))

# =========================================================
# 10. Save outputs
# =========================================================
sales_by_combo.to_csv("pareto_store_category_sales.csv", index=False)
pd.DataFrame({"store_category": pareto_combos}).to_csv("pareto_selected_combos.csv", index=False)
comparison_df.to_csv("gam_tuned_full_vs_pareto.csv", index=False)

print("\nSaved:")
print("- pareto_store_category_sales.csv")
print("- pareto_selected_combos.csv")
print("- gam_tuned_full_vs_pareto.csv")
```
